```python
from bs4 import BeautifulSoup
import re
import pandas as pd
from collections import OrderedDict
```

```python
with open("src/slutpriser/page1.html") as fp:
    soup = BeautifulSoup(fp)
```

```python
def keep_digits(text):
    text = re.sub(r'[^\d,]', '', text.strip())
    text = text.replace(',', '.')
    if not text:
        return -100
    return float(text)
```

```python
from lxml import html

with open("src/slutpriser/page1.html") as fp:
    tree = html.fromstring(fp.read())

def parse_xpath(xpath):
    return list(map(str.strip, tree.xpath(f'{xpath}/text()')))

pd.DataFrame(dict(
    address = parse_xpath('//*[@id="search-results"]/li[*]/div/div[1]/h2/span[2]'),
    area = parse_xpath('//*[@id="search-results"]/li[*]/div/div[1]/div/span[2]'),
    sqm = parse_xpath('//*[@id="search-results"]/li[*]/div/div[2]/div/div[1]'),
    rent = parse_xpath('//*[@id="search-results"]/li[*]/div/div[2]/div/div[2]'),
    price = parse_xpath('//*[@id="search-results"]/li[*]/div/div[3]/div[1]/span'),
    date = parse_xpath('//*[@id="search-results"]/li[*]/div/div[3]/div[2]/div[1]'),
    kr_sqm = parse_xpath('//*[@id="search-results"]/li[*]/div/div[3]/div[2]/div[2]'),
    #change = parse_xpath('//*[@id="search-results"]/li[*]/div/div[4]'),
))
```

```python
ul = soup.findAll('ul', {'class': 'sold-results'})[0]
listings = ul.findChildren('li', {'class': 'sold-results__normal-hit'}, recursive=False)

parsed_listings = []
for listing in listings:
    address = listing.findChildren(
        'h2', {'class': 'sold-property-listing__heading'}
    )[0].findChildren('span')[-1].text.strip()
    
    match = re.match(r'([A-ö0-9 ]*)(?=[ ,]+(\d+) ?tr)', address)
    if match:
        street, floor = match.groups()
    else:
        street, floor = address, -1
    
#     try:
#         loc = geolocator.geocode(street)
#         lat, lon = loc.latitude, loc.longitude
#     except:
#         lat = lon = 0

    stats = listing.findChildren('div', {'class': 'listing-card__attribute'})
    price, area, rooms, rent, kr_sqm = [stat.text.strip() for stat in stats][:5]
    data = OrderedDict((
        ('street', street),
        ('floor', int(floor)),
        ('price', keep_digits(price)),
        ('area', keep_digits(area)),
        ('rooms', keep_digits(rooms)),
        ('rent', keep_digits(rent)),
        ('kr_sqm', keep_digits(kr_sqm)),
        ('lat', lat),
        ('lon', lon),
        ))
    parsed_listings.append(data)

df = pd.DataFrame(parsed_listings)
```
